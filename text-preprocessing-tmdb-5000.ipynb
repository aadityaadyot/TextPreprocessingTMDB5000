```python
import re

def remove_emojis(text):
    # Remove emojis using regular expression
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
#Removing punctuations
import string

def remove_punctuations(data):
    for i in string.punctuation:
        if i in data:
            data = data.replace(i,"")
    return data
```

```python
data['overview'] = data['overview'].apply(remove_punctuations)
data.head(3)
```

```python
#removing stop words
import nltk
import os
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def stopwords_removal(data):
    token = nltk.word_tokenize(data)
    filter_word = [word for word in token if word not in stop_words]
    return ' '.join(filter_word)
```

```python
data['overview'] = data['overview'].apply(stopwords_removal)
data.head(3)
```

```python
# applying Stemming

from nltk.stem.porter import PorterStemmer

PS = PorterStemmer()
def stemming(data):
    stemm_word= [PS.stem(word) for word in data.split()]
    return ' '.join(stemm_word)
```

```python
data1['overview'] = data1['overview'].apply(stemming)
data1.head(3)
```

```python
#Applying Lemmatization
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet', force=True)
nltk.download('punkt', force=True)

WNL= WordNetLemmatizer()

def token_lemm(token):
    return WNL.lemmatize(token)

def lemmatizer_data(data):
    tokenized = nltk.word_tokenize(data)
    lemmatized_token = [token_lemm(token) for token in tokenized]
    return ' '.join(lemmatized_token)

data['overview'] = data['overview'].apply(lemmatizer_data)
data.head(3)
```
